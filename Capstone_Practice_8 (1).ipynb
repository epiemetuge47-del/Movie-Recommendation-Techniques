```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('movies.csv')
df1.head()
```

```python
df1.info()
```

```python
df2 = pd.read_csv('ratings.csv')
df2.head()
```

```python
df2.info()
```

```python
# Merge both data frames on movieid
df = pd.merge(df1, df2, on='movieId')
df.head()
```

```python
df.info()
```

```python
# Create User-Item Matrix (Hint: Use pandas pivot_table method with index = 'userId', columns = 'title', values = 'rating' )
df_pivot = pd.pivot_table(df, index='userId', columns='title', values='rating')
df_pivot.head()
```

```python
# Fill the row-wise NaNs in the User-Item Matrix with the corresponding user's mean ratings
df_pivot = df_pivot.apply(lambda row: row.fillna(row.mean()), axis=1)
df_pivot.head()
```

```python
corr_matrix = df_pivot.T.corr(method= 'pearson')
corr_matrix.head()
```

```python
# Choose the correlation of all users with only User 1 and sort in descending order
corr_user1 = corr_matrix[1]
corr_user1.head()
```

```python
# Sort the User 1 correlation in the descending order
corr_user1 = corr_user1.copy()
corr_user1.sort_values(ascending=False, inplace=True)
corr_user1.head()
```

```python
corr_user1.info()
```

```python
# Choose the top 50 users that are highly correlated to User 1
top_users = corr_user1[1:51].index
top_users
```

```python
df_movie32 = df[df['movieId'] == 32]
df_movie32.head()
```

```python
top_users_ratings_for_movie32 = df_movie32[df_movie32['userId'].isin(top_users)][['userId', 'rating']]
top_users_ratings_for_movie32.head()
```

```python
# Predict the rating that User 1 might give for the movie with movieid 32 based on the top 50 user correlation matrix.
# (Hint: Predicted rating = sum of [(weights) * (ratings)] / sum of (weights ). Here, weights is the correlation of the corresponding user with the first user). That is, the predicted rating is calculated as the weighted average of k similar users

# Create a temporary DataFrame to hold ratings and their corresponding correlations
temp_df = top_users_ratings_for_movie32.set_index('userId')
temp_df['correlation'] = corr_user1.loc[temp_df.index]

# Calculate the weighted sum of ratings
weighted_sum = (temp_df['rating'] * temp_df['correlation']).sum()

# Calculate the sum of weights
sum_of_weights = temp_df['correlation'].sum()

# Calculate the predicted rating
predicted_rating = weighted_sum / sum_of_weights

print(f"Predicted rating for User 1 for movie 32: {predicted_rating}")
```
