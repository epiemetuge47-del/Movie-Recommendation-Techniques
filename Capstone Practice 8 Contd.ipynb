```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv(r'C:\Users\Epie\Documents\movies.csv')
df1.head()
```

```python
df2 = pd.read_csv(r'C:\Users\Epie\Documents\ratings.csv')
df2.head()
```

```python
# Merge both data frames on movieid
df = pd.merge(df1, df2, on='movieId')
df.head()
```

```python
# Create User-Item Matrix (Hint: Use pandas pivot_table method with index = 'userId', columns = 'title', values = 'rating' )
df_pivot = pd.pivot_table(df, index='userId', columns='title', values='rating')
df_pivot.head()
```

```python
# Fill the row-wise NaNs in the User-Item Matrix with the corresponding user's mean ratings
df_pivot = df_pivot.apply(lambda row: row.fillna(row.mean()), axis=1)
df_pivot.head()
```

```python
# find Pearson correlation between movies
movie_correlation = df_pivot.corr(method='pearson')
movie_correlation.head()
```

```python
movie_correlation.isna().sum()
```

```python
# Choose the correlation of all movies with the movie Jurassic Park (1993) only
jurassic_park_corr = movie_correlation['Jurassic Park (1993)']
jurassic_park_corr.head()
```

```python
# Sort the Jurassic Park movie correlation in descending order
jurassic_park_corr_sorted = jurassic_park_corr.sort_values(ascending=False)
jurassic_park_corr_sorted.head()
```

```python
# Find 10 movies similar to the movie Jurassic Park (1993)
similar_movies = jurassic_park_corr_sorted[1:11]
similar_movies
```
